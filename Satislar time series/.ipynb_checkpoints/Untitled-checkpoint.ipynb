```python
from pandas.tseries.holiday import UKBankHolidayCalendar
import pandas as pd

cal = UKBankHolidayCalendar()
date = pd.to_datetime('2022-01-03')

if date in cal.holidays():
    print('{date} is a UK bank holiday'.format(date=date))
else:
    print('{date} is not a UK bank holiday'.format(date=date))
```
